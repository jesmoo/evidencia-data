```python
# Crear un diccionario

import pandas as pd

archivo_csv = '/app/data/aisles.csv'

df = pd.read_csv(archivo_csv)

aisle_dict = dict(zip(df['aisle_id'], df['aisle']))

print(aisle_dict)
```

```python
# Crear un análisis descriptivo 

import matplotlib.pyplot as plt

# Muestra las primeras filas del DataFrame para comprender los datos
print(f"Primeras filas del DataFrame: \n{df.head()}")
```

```python
# Descripción estadística de los datos numéricos

print(f"Descripción estadística de los datos numéricos:\n{df.describe()}")
```

```python
# Información general sobre el DataFrame
print(f"Información general sobre el DataFrame: {df.info()}")
```

```python
# Número de pasillos únicos

num_pasillos = df['aisle_id'].nunique()
print(f"Número de pasillos únicos: {num_pasillos}")
```

```python
# Los 10 pasillos más comunes
top_10_pasillos = df['aisle'].value_counts().head(10)

print(f"Los 10 pasillos más comunes: {top_10_pasillos}")
```

```python
# Gráfico de barras de los 10 pasillos más comunes

plt.figure(figsize=(12, 6))
plt.bar(top_10_pasillos.index, top_10_pasillos.values)
plt.xlabel('Pasillos')
plt.ylabel('Frecuencia')
plt.title('Los 10 Pasillos Más Comunes')
plt.xticks(rotation=90)
plt.show()
```
